```python
def open_files (file):
    """
    Open .json and .shape files and returns its geometries.
    
    Args:
        file (str) : Files with geoespatial information.
        
    Return:
        list of geometries
    
    """
    
    file = Path(file)
    if file.suffix == '.json':
        vector = open(file, mode='r', encoding='utf-8')
        data = json.load(vector)
        geometries = []
        for i in range(len(data['features'])):
            geometry = data['features'][i]['geometry']
            geometries.append(geometry)
        
    elif file.suffix == '.shp':
        geometries = [row.geometry for idx, row in gpd.read_file(file).iterrows()]
        
    
    return geometries
        
```

```python
def open_file(file):
    file = Path(file)
    gdf = gpd.read_file(file).to_crs('EPSG:32631') #project the coordenate system to raster CRS
    geometries = [row['geometry'] for i, row in gdf.iterrows()] #read all geometries in the file
    shapes = [g.buffer(50).envelope for g in geometries if g.type=='Point'] #buffer to points geometries (improve the visibility)
    shapes.extend([g for g in geometries if g.type=='Polygon']) #join all the geometries
    
    return shapes
    
```

```python
def composite(image_path, out_path, rgb=[4,3,2]):
    
    band_path = glob.glob(os.path.join(image_path, '*_B*.jp2'))
    bands = {int(a[a.find('_B')+2:a.find('_B')+4]):a for a in band_path}
    
    if len(list(set(bands.keys()) & set(rgb))) != 3:
        print(f'One or more input rgb({rgb}) bands not exists in source path')
        return
    
    b = rio.open(bands[rgb[2]], driver='JP2OpenJPEG') #blue
    g = rio.open(bands[rgb[1]], driver='JP2OpenJPEG') #green
    r = rio.open(bands[rgb[0]], driver='JP2OpenJPEG') #red
        
    #Empty raster write the composite
    emp = rio.open(out_path,'w',driver='Gtiff',
                        width=r.width, 
                        height=r.height, #width and height of any band
                        count=3,
                        crs=r.crs, #coordenate system
                        transform=r.transform,
                        dtype=r.dtypes[0])

    #combine the bands RGB in empty raster
    print('Making composite...')
    emp.write(b.read(1),3) #blue
    emp.write(g.read(1),2) #green
    emp.write(r.read(1),1) #red
#     ax = plt.subplots()
#     ep.plot_rgb(emp,
#             ax=ax,
#             title="Sentinel true color test and plot",
#             extent=emp)
    emp.close()
    print(f'Done in {out_path}')
```

```python
def plot_vector_and_raster (raster_path, vector_path): #raster_path .tiff; vector_path .shp
    
    # Set figure size and title size of plots
    mpl.rcParams['figure.figsize'] = (14, 14)
    mpl.rcParams['axes.titlesize'] = 20

    
    # Open polygons from shapefile with geopandas
    polygons = [row.geometry for idx, row in gpd.read_file(vector_path).iterrows()]
    #create the polygons features (edgecolor, line, fill, etc)
    patches = [PolygonPatch(feature, edgecolor="red", facecolor="none", linewidth=2) for feature in polygons]
    
    with rio.open(raster_path) as src:
        src_extent = plotting_extent(src)
        src_data = src.read()
        
    f, ax = plt.subplots()

    ep.plot_rgb(src_data,
                ax=ax,
                title="Sentinel true color test and plot",
                extent=src_extent)  # Use plotting extent from DatasetReader object

    # geometry.plot(ax=ax)
    ax.add_collection(mpl.collections.PatchCollection(patches, match_original=True))

    
    print(polygons[0].bounds) 

    return plt.show(), polygons
```

```python
def crop_raster_window (src, pol): 
    # Crop raster data to fire boundary
    src_data_crop, src_meta_crop = es.crop_image(src, (polygon for polygon in pol))
    

    # Define plotting extent using cropped array and transform from metadata
    src_crop_plot_extent = plotting_extent(src_data_crop[0], src_meta_crop["transform"])

     # Plot cropped data
            
    # Set figure size and title size of plots
    mpl.rcParams['figure.figsize'] = (14, 14)
    mpl.rcParams['axes.titlesize'] = 20

    f, ax = plt.subplots()

    x=  ep.plot_rgb(src_data_crop,
                    rgb=[0, 1, 2],
                    ax=ax,
                    title="Plot N 001-test",
                    extent=src_crop_plot_extent)  # Use plotting extent from cropped array

    ax.add_collection(mpl.collections.PatchCollection(patches, match_original=True))
   
    
    return plt.show()
```
