```python
out_path = '../Sentinelrgb.tiff'
in_path = '../Sentinelrgb.tiff'
```

```python
def rgb_composite (out_path, input_path, red, green, blue):
    #in an empty image write the composite
    rgb = rasterio.open(out_path,'w',driver='Gtiff',
                         width=red.width, height=red.height, #width and height of any band
                         count=3,
                         crs=red.crs, #coordenate system
                         transform=red.transform,
                         dtype=red.dtypes[0]
                         )
    rgb.write(blue.read(1),3) #blue
    rgb.write(green.read(1),2) #green
    rgb.write(red.read(1),1) #red
    rgb.close()

    # Open and Plot RGB with rasterio
    rgb = rasterio.open(input_path, count=3)
    plot.show(rgb)
```

```python
x = rgb_composite(out_path, in_path, band4, band3, band2)
```
